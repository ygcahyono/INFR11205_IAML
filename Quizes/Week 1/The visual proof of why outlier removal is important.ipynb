```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd

np.random.seed = 10
```

```python
arr = np.random.rand(100,1)
```

```python
arr.shape
```

```python
scaler = StandardScaler()
```

```python
arr_scale = scaler.fit_transform(arr)
```

```python
df = pd.DataFrame(arr_scale).reset_index()
df.columns = ['idx','cols']
```

```python
df.plot.scatter(x = 'idx', y = 'cols')
```

---

```python
randomise = np.random.default_rng()
```

```python
randomise_int = randomise.integers(1,10, size = (1000,1))
```

```python
arr_int_scale = scaler.fit_transform(randomise_int)
```

```python
arr_int_df = pd.DataFrame(arr_int_scale).reset_index()
arr_int_df.columns = ['idx','random']
```

```python
arr_int_df.plot.scatter('idx', y = 'random')
```

# Adding Uniform Distribution of Float Numbers

```python
arr = np.random.uniform(1.5, 75.5, size=(100,1))
```

```python
arr_df = pd.DataFrame(arr).reset_index()
arr_df.columns = ['idx', 'val']
```

```python
arr_df.plot.scatter('idx','val')
```

# Adding Uniform Distribution of Float Numbers (with Scaler)

```python
arr_scale = scaler.fit_transform(arr)
arr_df = pd.DataFrame(arr_scale).reset_index()
arr_df.columns = ['idx', 'val']
```

```python
arr_df.plot.scatter('idx','val')
```

# Adding Uniform Distribution of Float Numbers + Outlier 

```python
arr = np.append(arr, 1000)
```

```python
arr_df = pd.DataFrame(arr).reset_index()
arr_df.columns = ['idx', 'val']
```

```python
arr_df.plot.scatter('idx','val')
```

# Adding Uniform Distribution of Float Numbers + Outlier 

```python
arr = arr.reshape(-1,1)
```

```python
arr_scale = scaler.fit_transform(arr)
```

```python
arr_df = pd.DataFrame(arr_scale).reset_index()
arr_df.columns = ['idx','cols']
```

```python
arr_df.plot.scatter('idx','cols')
```
